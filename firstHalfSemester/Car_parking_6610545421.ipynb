```sql
CREATE TABLE customerInfo (
    customerID VARCHAR(255),
    LName VARCHAR(255),
    FName VARCHAR(255),
    phoneNo VARCHAR(255)
);
```

```sql
SELECT f.Fine
from fineInfo f;
```

```sql
-- avg revenue per customer
SELECT
    FName,
    AVG(f.Fine) as AVG
from customerInfo c , fineInfo f
GROUP BY C.FName 
```

```sql
-- Most Frequent customer
SELECT FName,
    count(customerID) as mostFrequent
from fineInfo , customerInfo
GROUP by FName
    
```
