```sql
CREATE TABLE CustomerDim
(
    CustomerKey INT NOT NULL,
    CustomerID INT NOT NULL,
    CustomerName VARCHAR(50) NOT NULL,
    CustomerAge INT NOT NULL,
    CustomerIncome DECIMAL(10,2) NOT NULL,
    CustomerCredit VARCHAR(50) NOT NULL,
    PRIMARY KEY (CustomerKey)
);
```

```sql
CREATE TABLE ProductDim
(
    ProductKey INT NOT NULL,
    ProductID INT NOT NULL,
    ProductName VARCHAR(50) NOT NULL,
    ProductCategory VARCHAR(50) NOT NULL,
    SupplierName VARCHAR(50) NOT NULL,
    ProductBrand VARCHAR(50) NOT NULL,
    ProductPrice DECIMAL(10,2) NOT NULL,
    ProductType VARCHAR(50) NOT NULL,
    ProductPlace_made VARCHAR(50) NOT NULL,
    PRIMARY KEY (ProductKey)
);
```

```sql
CREATE TABLE DateDim
(
    DateKey INT NOT NULL,
    Day INT NOT NULL,
    Month VARCHAR(50) NOT NULL,
    Year VARCHAR(4) NOT NULL,
    TheDate DATE NOT NULL,
    PRIMARY KEY (DateKey)
);
```

```sql
CREATE TABLE ElectFact
(
    ProductKey INT NOT NULL,
    CustomerKey INT NOT NULL,
    DateKey INT NOT NULL,
    unit_sold INT NOT NULL,
    total DECIMAL(10,2) NOT NULL,
    PRIMARY KEY (ProductKey, CustomerKey, DateKey),
    FOREIGN KEY (ProductKey) REFERENCES ProductDim(ProductKey),
    FOREIGN KEY (CustomerKey) REFERENCES CustomerDim(CustomerKey),
    FOREIGN KEY (DateKey) REFERENCES DateDim(DateKey)
);
```

For each product category, display total sale (with number of products-sold)

```sql
select ProductCategory , sum(total) as total_sum , sum(unit_sold) as total_product from ProductDim natural JOIN ElectFact GROUP BY ProductCategory ORDER by total_sum;
```

Show ALL customers (names and ages) and their total purchase (including those who havent bought anything)

```sql
select CustomerName , CustomerAge , sum(total) as totalSpent from CustomerDim c LEFT JOIN ElectFact e on c.CustomerKey = e.CustomerKey GROUP by CustomerAge , CustomerName ORDER by CustomerName , CustomerAge;
```

List customers with their most recent purchase date

```sql
SELECT CustomerName, CustomerIncome, CustomerCredit, MAX(TheDate) AS LastPurchasedDate FROM CustomerDim c LEFT JOIN (DateDim d LEFT JOIN ElectFact e ON d.DateKey = e.DateKey)  ON c.CustomerKey = e.CustomerKey GROUP BY CustomerName, CustomerIncome, CustomerCredit;
```
