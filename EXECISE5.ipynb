```sql
DROP DATABASE IF EXISTS EXECISE;
```

```sql
CREATE DATABASE EXECISE;
```

```sql
use EXECISE;
```

```sql
CREATE TABLE Person (
    PersonID INT PRIMARY KEY,
    pName VARCHAR(255)
);
```

```sql
CREATE TABLE Dealer (
    DealerID INT PRIMARY KEY,
    dName VARCHAR(255)
);
```

```sql
CREATE TABLE State_ (
    StateID INT PRIMARY KEY,
    sName VARCHAR(255)
);
```

```sql
CREATE TABLE Auto_ (
    AutoID INT PRIMARY KEY,
    aName VARCHAR(255),
    aModel VARCHAR(255),
    aYear VARCHAR(255),
    Own_PersonID INT,
    License_StateID INT,
    FOREIGN KEY (Own_PersonID) REFERENCES Person(PersonID)
)
```

```sql
CREATE TABLE ServiceBy (
    AutoID INT,
    DealerID INT,
    PRIMARY KEY (AutoID, DealerID),
    FOREIGN KEY (AutoID) REFERENCES Auto_(AutoID),
    FOREIGN KEY (DealerID) REFERENCES Dealer(DealerID)
);
```
